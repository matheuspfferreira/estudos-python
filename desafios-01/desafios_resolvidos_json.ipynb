***Desafio 03:*** Thais utilizou a função pd.read_json para carregar um arquivo JSON recebido em sua empresa. No entanto, ela notou que não conseguia acessar corretamente todos os dados do arquivo no DataFrame.

Um problema identificado foi que a coluna detalhes_compra apresentava informações importantes em formato de dicionário, dificultando a manipulação dos dados. Ajude Thais a acessar todas as colunas e informações disponíveis nos dados

```python
uri = "https://cdn3.gnarususercontent.com.br/4416-python-normalizando-arquivos-json/Projeto/Arquivos%20de%20atividades/questao_3.json"
df = pd.DataFrame(get(uri).json())
df.head()
```

```python
detalhes_compra = pd.json_normalize(df["detalhes_compra"])
detalhes_compra.head()
```

```python
df = pd.concat([df, detalhes_compra], axis=1)
df.head()
```

```python
df.drop(columns='detalhes_compra', inplace=True)
df.head()
```

***Desafio 04:*** Fábio recebeu um conjunto de dados relacionados às vendas da empresa onde trabalha. Ao inspecionar o arquivo, percebeu que se tratava de um JSON com uma estrutura aninhada de 3 níveis. O que Fábio pode fazer para transformar esse JSON em um conjunto de dados utilizável?

***Desafio 05:*** Tiago utilizou a normalização para transformar um arquivo JSON aninhado em um DataFrame no ambiente Python. Contudo, ele percebeu que o separador padrão ('.') nos nomes das colunas criadas a partir dos dados aninhados não atende às suas necessidades. Tiago gostaria de usar o caractere '_' como separador, resultando em nomes de colunas como cliente_id_cliente, cliente_nome_cliente, compra_id_produto e compra_nome_produto. Como Tiago pode alterar o separador de colunas para '_' ao realizar a normalização?

```python
uri = "https://cdn3.gnarususercontent.com.br/4416-python-normalizando-arquivos-json/Projeto/Arquivos%20de%20atividades/questao_5.json"
df = pd.DataFrame(get(uri).json())
df.head()
```

```python
clientes = pd.json_normalize(df["cliente"] , sep="_")
compras = pd.json_normalize(df["compra"], sep="_")
```

```python
df = pd.concat([df, clientes, compras], axis=1)
df.head()
```

```python
df.drop(columns=["cliente", "compra"], inplace=True)
df.head()
```

***Desafio 07:*** Juan recebeu um conjunto de dados em formato JSON e precisa normalizá-lo para acessar as informações de maneira estruturada. No entanto, ao utilizar json_normalize, ele identificou que o arquivo continha duas listas. Como Juan pode estruturar esses dados em um único DataFrame?

```python
uri = "https://cdn3.gnarususercontent.com.br/4416-python-normalizando-arquivos-json/Projeto/Arquivos%20de%20atividades/questao_7.json"
df = pd.DataFrame(get(uri).json())
df.head()
```

***Desafio 08:*** Paula está trabalhando com um conjunto de dados que contém duas listas aninhadas dentro de uma chave chamada lojas. Devido a essa estrutura, ela está enfrentando dificuldades para transformar o documento em um DataFrame manipulável. Ajude Paula a estruturar os dados em um DataFrame adequado para análise.

```python
uri = "https://cdn3.gnarususercontent.com.br/4416-python-normalizando-arquivos-json/Projeto/Arquivos%20de%20atividades/questao_8.json"
df = pd.DataFrame(get(uri).json())
df.head()
```

```python
nomes_lojas = df.index
df = pd.json_normalize(df["lojas"])
df.columns = nomes_lojas

df.head()
```

```python
df_nordeste = pd.json_normalize(df["lojas_nordeste"])
df_norte = pd.json_normalize(df["lojas_norte"])
```

```python
df = pd.concat([df_nordeste, df_norte], axis=0)
df.head()
```
